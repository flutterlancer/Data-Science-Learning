```python
# showing numpy logo as image
numpy_logo = np.load('../images/numpy logo.npy')

plt.imshow(numpy_logo, cmap='gray')
plt.title('Numpy Logo')
plt.show()
```

```python
# now convcerting the numpy logo into a dark version
dark_numpy_logo = 1 - numpy_logo
plt.imshow(dark_numpy_logo, cmap='gray')
plt.title('Numpy Logo')
plt.show()
```

##### **Other ways of inverting colors:**

```python
norm = (numpy_logo - numpy_logo.min()) / (numpy_logo.max() - numpy_logo.min())
inverted = 1.0 - norm
plt.imshow(inverted, cmap='gray')
plt.title('Manually Normalized & Inverted')
plt.show()
```

##### **Changing the background color only to dark grey:**

```python
dark_background_logo = numpy_logo.copy()
white_threshold = 0.95  # For normalized images (0–1 range)

# Create a mask for near-white pixels
if numpy_logo.max() <= 1.0:
    # Image is in 0–1 range
    white_pixels = np.all(numpy_logo > white_threshold, axis=-1)
    dark_background_logo[white_pixels] = [0.1, 0.1, 0.1]  # Dark gray
else:
    # Image is in 0–255 range
    white_pixels = np.all(numpy_logo > 240, axis=-1)
    dark_background_logo[white_pixels] = [30, 30, 30]  # Dark gray in 0–255

# Show the result
plt.imshow(dark_background_logo)
plt.title('numPy Logo with Dark Grey BG')
plt.show()
```
